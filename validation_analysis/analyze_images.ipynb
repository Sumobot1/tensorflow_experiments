```python
import os
import numpy as np
import matplotlib.pyplot as plt

FILE_PATH = "/hdd/Github/tensorflow_experiments/validation_results/01_10_2018__21_32_38_validation_run/epoch_0"
```

```python
x_val = np.load(os.path.join(FILE_PATH, "x_val.npy"))
y_val = np.load(os.path.join(FILE_PATH, "y_val.npy"))
y_pred_val = np.load(os.path.join(FILE_PATH, "y_pred_val.npy"))

x_val[0].shape
```

```python
# Cat should be [0,1], dog should be [1,0]
correct = np.equal(np.argmax(y_val, axis=1), np.argmax(y_pred_val, axis=1)).astype(np.float32)
avg = np.mean(correct)*100
# Is this correct?
w=x_val[0].shape[1]
h=x_val[1].shape[0]
fig=plt.figure(figsize=(80, 110))
columns = 4
rows = 5
pred_idx = 1
# y_val.shape[0]-np.sum(correct)
print("Avg: {}% correct.  Showing {} incorrect out of {}".format(avg, columns*rows, y_val.shape[0]))
for i in range(1, columns*rows +1):
    while np.argmax(y_val[pred_idx-1], axis=0) == np.argmax(y_pred_val[pred_idx - 1], axis=0):
        pred_idx += 1
    img = x_val[pred_idx-1]
    sub_plot = fig.add_subplot(rows, columns, i)
    sub_plot.set_xlabel('Predicted: [{:0.3f}, {:0.3f}] \n Actual:    [{:0.3f}, {:0.3f}]'.format(
        y_pred_val[pred_idx-1][0],
        y_pred_val[pred_idx-1][1],
        y_val[pred_idx-1][0],
        y_val[pred_idx-1][1]), fontsize=80)
    plt.imshow(img)
    plt
    pred_idx += 1
plt.show()
```
